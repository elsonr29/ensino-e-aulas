#### 1. Sobre a minimização de custos da firma, responda:  

Define a função de produção Cobb-Douglas, calcula as produtividades marginais, demonstra os retornos decrescentes e determina a TMST (itens a e b).

```python
import sympy as sp

# Definição de variáveis simbólicas
x1, x2 = sp.symbols('x1 x2', real=True, positive=True)
A, alpha, beta = sp.symbols('A alpha beta', real=True, positive=True)

# a) e b) Definição da Função de Produção Cobb-Douglas
f = A * (x1**alpha) * (x2**beta)

# Cálculo do Produto Marginal (Pmg)
pmg1 = sp.diff(f, x1)
pmg2 = sp.diff(f, x2)

# Demonstração de Produtos Marginais Decrescentes (Segunda Derivada < 0)
segunda_deriv_x1 = sp.diff(pmg1, x1)
segunda_deriv_x2 = sp.diff(pmg2, x2)

# Determinação da Taxa Marginal de Substituição Técnica (TMST)
tmst = pmg1 / pmg2

print(f"Produto Marginal x1: {sp.simplify(pmg1)}")
print(f"Produto Marginal x2: {sp.simplify(pmg2)}")
print(f"Segunda Derivada x1: {sp.simplify(segunda_deriv_x1)} (Negativa se alpha < 1)")
print(f"TMST 1,2: {sp.simplify(tmst)}")
```

#### Formalização de Isocusto e Otimização (itens c e d)

O código abaixo define a reta de isocusto e encontra analiticamente o ponto de minimização de custo (tangência).

```python
w1, w2, y_alvo = sp.symbols('w1 w2 y_alvo', real=True, positive=True)

# c) Reta de Isocusto (Isolando x2)
# w1*x1 + w2*x2 = C -> x2 = C/w2 - (w1/w2)*x1
inclinação_isocusto = w1 / w2

# d) Ponto de Minimização (TMST = Razão de Preços)
condicao_otimo = sp.Eq(tmst, w1 / w2)
relacao_x1_x2 = sp.solve(condicao_otimo, x2)[0]

print(f"Relação de substituição ótima: x2 = {relacao_x1_x2}")
```

#### Choques e gráficos (itens e e f)

Gera a representação gráfica completa. Demonstra a isoquanta original, o equilíbrio inicial e as mudanças provocadas por um aumento no preço de x1​ (Efeito Substituição) ou por um aumento na produção.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parâmetros da tecnologia Cobb-Douglas
A, a, b = 1, 0.3, 0.7
y_alvo = 100
w1_inicial, w2 = 10, 6
w1_novo = 15

def calcular_equilibrio(w1, w2, y):
    # Condição de tangência: (a/b) * (x2/x1) = w1/w2
    # x2 = (w1/w2) * (b/a) * x1
    # Substituindo na função de produção: y = A * x1^a * [(w1/w2)*(b/a)*x1]^b
    termo_precos = (w1 / w2) * (b / a)
    x1_otimo = (y / (A * (termo_precos**b)))**(1/(a + b))
    x2_otimo = termo_precos * x1_otimo
    custo_minimo = w1 * x1_otimo + w2 * x2_otimo
    return x1_otimo, x2_otimo, custo_minimo

# Cálculos para os dois cenários
x1_ini, x2_ini, c_ini = calcular_equilibrio(w1_inicial, w2, y_alvo)
x1_pos, x2_pos, c_pos = calcular_equilibrio(w1_novo, w2, y_alvo)

def isoquanta(x1, y):
    return (y / (A * x1**a))**(1/b)

x_space = np.linspace(5, 120, 500)

plt.figure(figsize=(10, 7))

# Plotagem da Isoquanta fixa em 100 unidades
plt.plot(x_space, isoquanta(x_space, 100), label='Isoquanta y=100', color='blue', lw=2.5)

# Plotagem das Isocustos Tangentes calculadas
# Isocusto: x2 = c/w2 - (w1/w2)*x1
plt.plot(x_space, c_ini/w2 - (w1_inicial/w2)*x_space, color='darkgreen', 
         linestyle='--', alpha=0.7, label=f'Isocusto Inicial (c={c_ini:.1f})')

plt.plot(x_space, c_pos/w2 - (w1_novo/w2)*x_space, color='darkred', 
         linestyle='--', alpha=0.7, label=f'Isocusto Novo w1 (c={c_pos:.1f})')

# Marcação dos pontos de escolha ótima
plt.scatter(x1_ini, x2_ini, color='green', s=100, zorder=5, label='Ótimo Inicial')
plt.scatter(x1_pos, x2_pos, color='red', s=100, zorder=5, label='Ótimo com w1 alto')

plt.title('Minimização de Custos: Representação da Tangência')
plt.xlabel('Insumo x1')
plt.ylabel('Insumo x2')
plt.ylim(0, 350)
plt.xlim(0, 120)
plt.legend()
plt.grid(True, alpha=0.2)
plt.show()
```

2. Considere a função de produção e a estrutura de custos de uma firma de tecnologia. A empresa utiliza dois insumos: capital tecnológico ($K$), representado por agentes de inteligência artificial; e trabalho ($L$), representado por programadores. Os custos mensais por unidade de fator correspondem, respectivamente, à taxa de locação da IA ($w_K=10$) e ao salário ($w_L=6$). Com base nestas informações, responda aos itens a seguir:  

$$		\text{Função de produção:} \quad f(K, L) = K^{0,3} L^{0,7} = 100$$
	
$$	\text{Restrição orçamentária:} \quad 10K + 6L = c $$


```python
import sympy as sp
import pandas as pd

def resolver_minimizacao_custos():
    # 1. Definição das variáveis simbólicas
    K, L = sp.symbols('K L', real=True, positive=True)
    wK, wL, y_alvo = sp.symbols('wK wL y_alvo', real=True, positive=True)

    # 2. Definição da função de produção Cobb-Douglas
    # f(K, L) = K^0.3 * L^0.7
    alpha = 0.3
    beta = 0.7
    f = K**alpha * L**beta

    # 3. Cálculo das Produtividades Marginais (Pmg)
    pmgK = sp.diff(f, K)
    pmgL = sp.diff(f, L)

    # 4. Taxa Marginal de Substituição Técnica (TMST = PmgK / PmgL)
    tmst = pmgK / pmgL

    # 5. Condição de Otimalidade: TMST = wK / wL
    # Resolvendo a condição para expressar L em termos de K
    condicao_otimo = sp.Eq(tmst, wK / wL)
    relacao_L = sp.solve(condicao_otimo, L)[0]

    # 6. Substituição na restrição tecnológica para encontrar K*
    equacao_k = sp.Eq(K**alpha * relacao_L**beta, y_alvo)
    
    # Função para calcular os valores numéricos para cada cenário
    def calcular_cenario(p_wK, p_wL, p_y):
        # Substitui os parâmetros na equação de K e resolve
        sol_k = sp.solve(equacao_k.subs({wK: p_wK, wL: p_wL, y_alvo: p_y}), K)[0]
        # Calcula L com base no K encontrado
        sol_l = relacao_L.subs({K: sol_k, wK: p_wK, wL: p_wL})
        # Calcula o custo total
        custo = p_wK * sol_k + p_wL * sol_l
        
        return {
            "wK": p_wK,
            "wL": p_wL,
            "Produção (y)": p_y,
            "Capital (K*)": round(float(sol_k), 2),
            "Trabalho (L*)": round(float(sol_l), 2),
            "Custo Total (c)": round(float(custo), 2)
        }

    # Execução dos cenários solicitados
    resultados = []
    
    # a) Escolha ótima inicial
    resultados.append(calcular_cenario(10, 6, 100))
    
    # b) Choque no preço de K (wK' = 4)
    resultados.append(calcular_cenario(4, 6, 100))
    
    # c) Choque no preço de L (wL' = 3) a partir de (b)
    resultados.append(calcular_cenario(4, 3, 100))
    
    # d) Aumento da produção para y = 200 (com preços do item a)
    resultados.append(calcular_cenario(10, 6, 200))

    # Formatação dos resultados em um DataFrame para visualização
    df = pd.DataFrame(resultados)
    df.index = ['Item (a)', 'Item (b)', 'Item (c)', 'Item (d)']
    
    return df

# Executar a resolução
tabela_resultados = resolver_minimizacao_custos()
print(tabela_resultados)
```

```python
# gráficos conjuntos

import numpy as np
import matplotlib.pyplot as plt

def plotar_otimizacao(df):
    # Configurações de estilo
    plt.figure(figsize=(12, 8))
    k_vizinhança = np.linspace(5, 200, 400)
    
    # Função para calcular a Isoquanta: L = (y / K^alpha)^(1/beta)
    def isoquanta(K_vals, y, a=0.3, b=0.7):
        return (y / (K_vals**a))**(1/b)

    # Cores para distinguir os cenários
    cores = ['blue', 'green', 'red', 'purple']
    labels = df.index
    
    for i, cenario in enumerate(labels):
        k_otimo = df.loc[cenario, "Capital (K*)"]
        l_otimo = df.loc[cenario, "Trabalho (L*)"]
        y_val = df.loc[cenario, "Produção (y)"]
        wk_val = df.loc[cenario, "wK"]
        wl_val = df.loc[cenario, "wL"]
        custo_val = df.loc[cenario, "Custo Total (c)"]
        
        # Plotagem da Isoquanta
        plt.plot(k_vizinhança, isoquanta(k_vizinhança, y_val), 
                 color=cores[i], alpha=0.3, label=f'Isoquanta y={y_val}')
        
        # Plotagem da Reta de Isocusto: L = c/wL - (wK/wL)*K
        isocusto = custo_val/wl_val - (wk_val/wl_val)*k_vizinhança
        plt.plot(k_vizinhança, isocusto, color=cores[i], 
                 linestyle='--', alpha=0.5, label=f'Isocusto {cenario}')
        
        # Marcação do Ponto Ótimo
        plt.scatter(k_otimo, l_otimo, color=cores[i], s=100, zorder=5)
        plt.annotate(f'({k_otimo}, {l_otimo})', (k_otimo, l_otimo), 
                     textcoords="offset points", xytext=(0,10), ha='center', fontweight='bold')

    # Ajustes finais do gráfico
    plt.title('Minimização de Custos: Isoquantas e Isocustos', fontsize=14)
    plt.xlabel('Capital (K)', fontsize=12)
    plt.ylabel('Trabalho (L)', fontsize=12)
    plt.xlim(0, 200)
    plt.ylim(0, 400)
    plt.grid(True, which='both', linestyle=':', alpha=0.5)
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()

# Chamada da função com o DataFrame gerado anteriormente
plotar_otimizacao(tabela_resultados)
```

```python
# gráficos individuais

import numpy as np
import matplotlib.pyplot as plt

# Configuração global de plotagem
def configurar_grafico(titulo, x_max=250, y_max=400):
    plt.figure(figsize=(8, 6))
    plt.xlim(0, x_max)
    plt.ylim(0, y_max)
    plt.xlabel('Capital (K)')
    plt.ylabel('Trabalho (L)')
    plt.title(titulo, fontsize=12, fontweight='bold')
    plt.grid(True, linestyle=':', alpha=0.6)

def plotar_isoquanta(y, alpha, beta, cor, label):
    K_vals = np.linspace(5, 300, 500)
    L_vals = (y / (K_vals**alpha))**(1/beta)
    plt.plot(K_vals, L_vals, color=cor, lw=2, label=label)

def plotar_isocusto(custo, wK, wL, cor, label):
    K_vals = np.linspace(0, 300, 500)
    L_vals = custo/wL - (wK/wL)*K_vals
    plt.plot(K_vals, L_vals, color=cor, linestyle='--', alpha=0.7, label=label)

def marcar_ponto(K, L, cor):
    plt.scatter(K, L, color=cor, s=80, zorder=5)
    plt.annotate(f'({K:.1f}, {L:.1f})', (K, L), xytext=(5, 5), textcoords='offset points', fontweight='bold')
```

3. O choque tecnológico altera a função de produção da firma de tecnologia, resultando em uma maior participação do capital tecnológico ($K$). Com base nessa mudança estrutural, responda aos itens a seguir:

$$\text{Função de produção (após choque):} \quad f(K, L) = K^{0,6} L^{0,4} = 100$$

$$\text{Restrição orçamentária:} \quad 10K + 6L = c$$


```python
import sympy as sp
import pandas as pd

def resolver_mudanca_tecnologica():
    # 1. Definição das variáveis
    K, L = sp.symbols('K L', real=True, positive=True)
    wK, wL, y_alvo = 10, 6, 100
    
    # Parâmetros da Questão 2 (Anterior) e Questão 3 (Nova)
    tecnologias = {
        "Q2 (Original)": {"alpha": 0.3, "beta": 0.7},
        "Q3 (Nova)": {"alpha": 0.6, "beta": 0.4}
    }
    
    resultados = []

    for nome, params in tecnologias.items():
        a, b = params["alpha"], params["beta"]
        f = K**a * L**b
        
        # Cálculo da TMST (PmgK / PmgL)
        pmgK = sp.diff(f, K)
        pmgL = sp.diff(f, L)
        tmst = pmgK / pmgL
        
        # Condição de Otimalidade: TMST = wK / wL
        condicao = sp.Eq(tmst, wK / wL)
        relacao_L = sp.solve(condicao, L)[0]
        
        # Substituição na isoquanta de produção (y=100)
        equacao_k = sp.Eq(K**a * relacao_L**b, y_alvo)
        sol_k = sp.solve(equacao_k, K)[0]
        sol_l = relacao_L.subs(K, sol_k)
        custo = wK * sol_k + wL * sol_l
        
        resultados.append({
            "Cenário": nome,
            "Alpha (IA)": a,
            "Beta (Prog)": b,
            "Capital (K*)": round(float(sol_k), 2),
            "Trabalho (L*)": round(float(sol_l), 2),
            "Custo Total (c)": round(float(custo), 2)
        })

    return pd.DataFrame(resultados).set_index("Cenário")

# Execução
df_comparativo = resolver_mudanca_tecnologica()
print(df_comparativo)
```

```python
# Comparação entre as funções de produção (original e nova), mas mantendo preços e produção constantes

import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# 1. Resolução Analítica com SymPy
K, L = sp.symbols('K L', real=True, positive=True)
wK_fixo, wL_fixo, y_alvo = 10, 6, 100

def encontrar_equilibrio(alpha, beta):
    f = K**alpha * L**beta
    tmst = sp.diff(f, K) / sp.diff(f, L)
    
    # Condição de tangência: TMST = wK / wL
    condicao = sp.Eq(tmst, wK_fixo / wL_fixo)
    relacao_L = sp.solve(condicao, L)[0]
    
    # Substituição na isoquanta y = 100
    eq_prod = sp.Eq(K**alpha * relacao_L**beta, y_alvo)
    k_res = float(sp.solve(eq_prod, K)[0])
    l_res = float(relacao_L.subs(K, k_res))
    custo_res = wK_fixo * k_res + wL_fixo * l_res
    
    return k_res, l_res, custo_res

# Cálculo dos dois cenários
k1, l1, c1 = encontrar_equilibrio(0.3, 0.7) # Tecnologia Inicial
k2, l2, c2 = encontrar_equilibrio(0.6, 0.4) # Tecnologia Nova

# 2. Visualização Gráfica
plt.figure(figsize=(10, 7))
k_plot = np.linspace(10, 250, 500)

# Tecnologia Inicial (Azul)
plt.plot(k_plot, (y_alvo / (k_plot**0.3))**(1/0.7), color='blue', label='Tecnologia Inicial (α=0.3)', lw=2)
plt.plot(k_plot, c1/wL_fixo - (wK_fixo/wL_fixo)*k_plot, color='blue', linestyle=':', alpha=0.4)
plt.scatter(k1, l1, color='blue', s=100, zorder=5)

# Nova Tecnologia (Laranja)
plt.plot(k_plot, (y_alvo / (k_plot**0.6))**(1/0.4), color='orange', label='Nova Tecnologia (α=0.6)', lw=2)
plt.plot(k_plot, c2/wL_fixo - (wK_fixo/wL_fixo)*k_plot, color='orange', linestyle=':', alpha=0.4)
plt.scatter(k2, l2, color='orange', s=100, zorder=5)

# Ajustes do Gráfico
plt.title('Impacto da Mudança Tecnológica (Preços Fixos: wK=10, wL=6)', fontsize=12)
plt.xlabel('Capital Tecnológico (K)')
plt.ylabel('Trabalho (L)')
plt.xlim(0, 250)
plt.ylim(0, 300)
plt.grid(True, linestyle=':', alpha=0.6)
plt.legend()
plt.show()

print(f"Equilíbrio Q2: K={k1:.2f}, L={l1:.2f}, Custo={c1:.2f}")
print(f"Equilíbrio Q3: K={k2:.2f}, L={l2:.2f}, Custo={c2:.2f}")
```
